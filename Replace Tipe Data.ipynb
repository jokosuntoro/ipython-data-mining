```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('../pengukuran.csv')
```

```python
data['distance']=data['distance'].str.replace(',','.').astype(float)
data['consume']=data['consume'].str.replace(',','.').astype(float)
data['temp_inside']=data['temp_inside'].str.replace(',','.').astype(float)
data
```

```python
data_filter = data[['distance', 'speed']]
```

```python
#Univariate analysis distance
#Melihat distribusi dari distance
# kind = hist/kde
fig = plt.figure(figsize=(12,4))

fig.add_subplot(1,2,1)
data_filter['distance'].plot(kind='kde')

fig.add_subplot(1,2,2)
plt.boxplot(data_filter['distance'])
plt.show()
```

```python
#Univariate analysis speed
#Melihat distribusi dari speed
# kind = hist/kde
fig = plt.figure(figsize=(12,4))

fig.add_subplot(1,2,1)
data_filter['speed'].plot(kind='kde')

fig.add_subplot(1,2,2)
plt.boxplot(data_filter['speed'])
plt.show()
```

```python
# Bivariate analysis atribut distance dan speed
#Menggunakan scatter plot
plt.scatter(data['distance'], data['speed'])
plt.xlabel('distance')
plt.ylabel('speed')
plt.title('Scatter Plot Atribut Distance vs Speed')
plt.show()
```
