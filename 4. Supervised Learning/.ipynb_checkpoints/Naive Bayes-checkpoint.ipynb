```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
data = pd.read_csv('../datasets/golf.csv')
```

```python
data.shape
```

```python
data.head(14)
```

```python
encoder = LabelEncoder()
```

```python
data['cuaca'] = encoder.fit_transform(data['cuaca'].values)
data['angin'] = encoder.fit_transform(data['angin'].values)
```

```python
data.head(14)
```

```python
atribut = data.drop(columns='kelas')
```

```python
atribut.head(14)
```

```python
kelas = data['kelas']
```

```python
kelas.head(14)
```

```python
nb = GaussianNB()
nb.fit(atribut, kelas)
```
